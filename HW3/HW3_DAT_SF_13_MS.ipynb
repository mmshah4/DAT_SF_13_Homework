```python
#converting other categorican values to numerical values
# convert car to 0 and truck to 1
bank['education'] = bank.education.map({'primary':0,'secondary':1,'tertiary':2,'unknown':3})
bank['marital'] = bank.marital.map({'single':0,'married':1,'divorced':2})
bank['default'] = bank.default.map({'yes':0,'no':1})
bank['housing'] = bank.housing.map({'yes':0,'no':1})
bank['loan'] = bank.loan.map({'yes':0,'no':1})
bank['contact'] = bank.contact.map({'cellular':0,'unknown':1,'telephone':2})
bank['month'] = bank.month.map({'apr':0,'aug':1,'dec':2,'feb':3,'jan':4,'jul':5,'jun':6,'mar':7,'may':8,'nov':9,'oct':10,'sep':11})
bank['job'] = bank.job.map({'admin.':0, 'blue-collar':1, 'entrepreneur':2, 'housemaid':3, 'management':4, 'retired':5, 'self-employed':6, 'services':7, 'student':8, 
                                    'technician':9, 'unemployed':10, 'unknown':11})
bank['poutcome'] = bank.poutcome.map({'unknown':0,'failure':1,'other':2,'success':3})
bank['y'] = bank.y.map({'yes':0,'no':1})
```

```python
categorical_features = ['job', 'marital', 'education', 'default','housing', 'loan', 'contact', 'month', 'poutcome']
numeric_features = ['age','balance','day','duration','campaign','pdays','previous']
target = bank.y
cat_bank = bank[categorical_features]
num_bank = bank[numeric_features]
```

```python
#converting pandas dataframe to numpy matrix
np_cat_bank = cat_bank.as_matrix()
```

```python
#converting all categorical features into dummy variables
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder()
enc.fit(np_cat_bank)  
OneHotEncoder(categorical_features='all',n_values='auto', sparse=True)
enc.n_values_
```

```python
enc.feature_indices_
```

```python
#Transforming the categorical matrix
tr_cat_bank = enc.transform(np_cat_bank).toarray()
```

```python
#Merging categorical and numerical arrays
merge_bank =np.hstack((tr_cat_bank, num_bank.as_matrix()))
```

```python
#check if the stacking is done correctly
print tr_cat_bank[1],num_bank.as_matrix()[1], merge_bank[1]
```

```python
#Building a decision tree
# split into train/test
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(merge_bank, target, random_state=1)
```

```python
# print out each of the arrays
print X_train
print y_train
print X_test
print y_test
```

```python
# fit a classification tree with max_depth=3 on all data
from sklearn.tree import DecisionTreeClassifier
from sklearn.cross_validation import cross_val_score
treeclf = DecisionTreeClassifier(max_depth=20, random_state=1)
treeclf.fit(X_train, y_train)

```

```python
# make predictions
preds = treeclf.predict(X_test)

# print mean of predictions and actual values
print preds.mean(), y_test.mean()
```

```python
#Cross validation 5 fold
depth=40
scores=[]
scores1=[]
for i in range(depth):
    treeclf = DecisionTreeClassifier(max_depth=i+2, random_state=1)
    scores = cross_val_score(treeclf, merge_bank, target, cv=5)
    scores1.append(scores)
```

```python
from seaborn import plt
import matplotlib.pyplot as plt
scores2=[]
for i in range(depth):
    new = scores1[i].mean()
    scores2.append(new)
from matplotlib.colors import ListedColormap
plt.plot(range(depth), scores2, linewidth=3.0)
```
