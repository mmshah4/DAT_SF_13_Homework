```python
# setting up an array to call out nearest neighbors from 1 through 150
n_neighbors = range(1, 100,1)
print n_neighbors
```

```python
# Create the training (and test) set using scikit-learn's train_test_split function
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3, random_state=12)
```

```python
print X_train[:5]
```

```python
# Loop through each neighbors value from 1 to 150 and append
# the scores
scores = []
for n in n_neighbors:
    clf = neighbors.KNeighborsClassifier(n)
    clf.fit(X_train, y_train)
    scores.append(clf.score(X_test, y_test))
```

```python
plt.plot(n_neighbors, scores, linewidth=3.0)
for i in len(scores):
    print i, scores[i]
```

```python
#Cross validation 5 fold
from sklearn.cross_validation import cross_val_score
scores=[]
scores1=[]
for i in n_neighbors:
    clf = neighbors.KNeighborsClassifier(i, weights='uniform')
    clf.fit(iris.data, iris.target)
    scores = cross_val_score(clf, iris_df.values[:,:4], iris.target, cv=5)
    scores1.append(scores)
```

```python
#splitting the datasets into equal splits based on the number of CV we would like to perform
import math
folds=5
total_cnt=len(iris_df)
n_n = 50
sample_cnt = math.floor(total_cnt/(folds-1))-1
rndm_series = np.random.permutation(np.arange(total_cnt))
#print len(rndm_series), len(iris_df)

iris_df_random=iris_df.ix[rndm_series]
j=len(iris_df_random)

iris_df_random['rndm_index'] = iris_df_random.index
iris_df_random.reset_index(level=0, inplace=True)

iris_df_random['sample_cnt1'] = sample_cnt
iris_df_random['new_index'] = iris_df_random.index
iris_df_random['sample2'] = np.round(iris_df_random['new_index']/iris_df_random['sample_cnt1'],0)

scores_final=[]
for n in range(1,50):
    scores=[]
    for i in range(0,folds):
        X_train = iris_df_random[iris_df_random['sample2'] != i][['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']].values
        Y_train = iris_df_random[iris_df_random['sample2'] != i]['Target'].values
        X_test = iris_df_random[iris_df_random['sample2'] == i][['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']].values
        Y_test = iris_df_random[iris_df_random['sample2'] == i]['Target'].values
        clf = neighbors.KNeighborsClassifier(n)
        clf.fit(X_train,Y_train)
        scores.append(clf.score(X_test, Y_test))
    scores_final.append(scores)
```

```python
type(scores)
scores_array = np.array(scores_final)
type(scores_array)
```

```python
summary=[]
for i in range(1,40):
    new = scores_array[i].mean()
    summary.append(new)
print summary
```
